```python
# Delete the column education and the number of open complaints from the dataframe.
tit_data = tit_data.drop(['education', 'number_of_open_complaints'], axis = 1)
tit_data.columns
```

```python
def remove_percent(x: str):
    return x.replace('%', '')
```

```python
tit_data['customer_lifetime_value'] = tit_data['customer_lifetime_value'].astype(str).apply(remove_percent).astype(float)
tit_data['customer_lifetime_value']
```

```python
def multiply_100(x: float):
    return x*100

multiply_100(1.20)
```

```python
tit_data['customer_lifetime_value'] = tit_data['customer_lifetime_value'].apply(multiply_100)
tit_data['customer_lifetime_value']
```
