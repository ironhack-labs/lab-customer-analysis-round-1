```python
#Read the three files into python as dataframes

import pandas as pd

df1 = pd.read_csv(r"C:\Users\pedro\Desktop\Ironhack\lab-customer-analysis-round-1\files_for_lab\csv_files\file1.csv")
df2 = pd.read_csv(r"C:\Users\pedro\Desktop\Ironhack\lab-customer-analysis-round-1\files_for_lab\csv_files\file2.csv")
df3 = pd.read_csv(r"C:\Users\pedro\Desktop\Ironhack\lab-customer-analysis-round-1\files_for_lab\csv_files\file3.csv")

df1
df2
df3
```

```python
#Standardize header names.

cols = []
for i in range(len(df1.columns)):
    cols.append(df1.columns[i].lower().replace(' ','_'))

df1.columns = cols
df1
```

```python
cols = []
for i in range(len(df2.columns)):
    cols.append(df2.columns[i].lower().replace(' ','_'))

df2.columns = cols
df2
```

```python
cols = []
for i in range(len(df3.columns)):
    cols.append(df3.columns[i].lower().replace(' ','_'))

df3.columns = cols
df3
```

```python
#Rearrange the columns in the dataframe as needed

df1 = df1.rename(columns={'st':'state'})
df1
```

```python
df2 = df2.rename(columns={'st':'state'})
df2
```

```python
#Concatenate the three dataframes

df_all = pd.concat([df1,df2,df3], axis=0)

df_all
```

```python
df_all.shape
```

```python
'''
Which columns are numerical?
income, monthly_premium_auto, total_claim_amount 
'''

df_all.dtypes
```

```python
#Delete the column education and the number of open complaints from the dataframe.

df_all = df_all.drop(['education','number_of_open_complaints'], axis=1)
df_all
```

```python
'''
Correct the values in the column customer lifetime value. 
They are given as a percent, so multiply them by 100 and change dtype to numerical type.
'''
df_all['customer_lifetime_value'] = pd.to_numeric(df_all['customer_lifetime_value'].replace("[%,]","", regex=True))
df_all

df_all['customer_lifetime_value']=df_all['customer_lifetime_value']*100
df_all

df_all.dtypes
```

```python
#Check for duplicate rows in the data and remove if any.

df_all.duplicated()

df_all.drop_duplicates() 
```
