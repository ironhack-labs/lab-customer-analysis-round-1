#### Standardize header names and Rearrange the columns in the dataframe as needed.

```python
cols = []
for column in range(len(file1.columns)):
    cols.append(file1.columns[column].upper())
file1.columns = cols
```

```python
cols = []
for column in range(len(file2.columns)):
    cols.append(file2.columns[column].upper())
file2.columns = cols
```

```python
cols = []
for column in range(len(file3.columns)):
    cols.append(file3.columns[column].upper())
file3.columns = cols
```

```python
file1 = file1.rename(columns={'ST':'STATE'})
file2 = file2.rename(columns={'ST':'STATE'})
```

```python
file1=file1[['CUSTOMER', 'STATE', 'CUSTOMER LIFETIME VALUE', 'EDUCATION', 'GENDER',
       'INCOME', 'MONTHLY PREMIUM AUTO', 'NUMBER OF OPEN COMPLAINTS',
       'POLICY TYPE', 'TOTAL CLAIM AMOUNT', 'VEHICLE CLASS']]
```

```python
file2=file2[['CUSTOMER', 'STATE', 'CUSTOMER LIFETIME VALUE', 'EDUCATION', 'GENDER',
       'INCOME', 'MONTHLY PREMIUM AUTO', 'NUMBER OF OPEN COMPLAINTS',
       'POLICY TYPE', 'TOTAL CLAIM AMOUNT', 'VEHICLE CLASS']]
```

#### Concatenate the three dataframes

```python
data = pd.concat([file1,file2,file3])
data
```

#### Which columns are numerical?

```python
data.select_dtypes('number')
```

The columns that are numerical are Income, Monthly Premium Auto and Total Claim Amount

#### Which columns are categorical?

```python
data.select_dtypes('object')
```

The columns that are numerical are Customer, State, Customer Lifetime Value, Education, Gender, Number of Open Complaints, Policy Type and Vehicle Class

#### Delete the column education and the number of open complaints from the dataframe.


```python
data.drop(['EDUCATION','NUMBER OF OPEN COMPLAINTS'],axis=1,inplace=True)
```

#### Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.

```python
data
```

```python
data['CUSTOMER LIFETIME VALUE']=data['CUSTOMER LIFETIME VALUE'].astype(str).str.replace('%',"").astype(float) / 100
```

#### Check for duplicate rows in the data and remove if any.

```python
data.drop_duplicates()
```
